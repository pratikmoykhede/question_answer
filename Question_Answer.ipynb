```python
import pandas as pd
```

```python
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)

questions=['Why are the names ?','what are the ages?']
context='The names are swara and pratik.Ages are 16 and 25 '

result=[]

for question in questions:

  QA_input = {
      'question':question,
      'context':context
  }
  res = nlp(QA_input)
  result.append(res)

  # print(result)

df=pd.DataFrame({
    'Questions':questions,
    'Answer':[x['answer'] for x in result]
})

print(df)
```
